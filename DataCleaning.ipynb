# Data Frame


```python
import pandas as pd 
import numpy as np

# dataframe using dictionary 
country = ['INDIA', 'CHINA', 'RUSSIA', 'USA', 'JAPAN']
age = [55, 45,70,65,80]
salary = [55000, 45000,90000,100000,150000]
data_frame = pd.DataFrame({'Country': country, 'Age': age, 'Salary': salary})
data_frame
```

```python
purchase = [20000, 150000,30000,40000,50000]
data_frame1 = pd.DataFrame(purchase,columns=['purchase'])
data_frame1
```

```python
data_frame2 = pd.DataFrame({'Country': country, 'Age': age, 'Salary': salary, 'Purchase':purchase})
data_frame2
```

```python
# still do merging
data_frame3 = pd.DataFrame({'Purchase':purchase, 'Country':country})
data_frame3 
```

```python
country_data = data_frame.merge(data_frame3 , how ='inner', on='Country')
country_data 
```

```python
#download csv file
country_data.to_csv('country_dataset.csv', index=False)
```
